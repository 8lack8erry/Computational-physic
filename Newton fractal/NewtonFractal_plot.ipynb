```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("NewtonFractal.dat", delimiter="\t")

x = data["Re(z)"]
y = data["Im(z)"]
z = data["n"]

fig, ax = plt.subplots()

scatter = ax.scatter(x, y, c=z, cmap='viridis')
fig.colorbar(scatter, ax=ax, label='n')

plt.xlabel('Re(z)')
plt.ylabel('Im(z)')

plt.tight_layout()
plt.savefig("NewtonFractal_depth.png")
plt.show()
```
