```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data_Esercizio20.dat", delimiter="\t")

x = data["Re(z)"]
y = data["Im(z)"]
z = data["n"]

fig, ax = plt.subplots()

scatter = ax.scatter(x, y, c=z, cmap='viridis')
fig.colorbar(scatter, ax=ax, label='n')

plt.xlabel('Re(z)')
plt.ylabel('Im(z)')

plt.tight_layout()
plt.savefig("scatter_plot.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data_Es20_no0.dat", delimiter="\t")

x = data["Re(z)"]
y = data["Im(z)"]
z = data["n"]
r = data["theta"]
precision = 1e-4
zeros = [0, 2. * np.pi / 3, -2. * np.pi / 3]

fig, ax = plt.subplots(figsize=(10, 5))

x1, x2, x3 = [], [], []
y1, y2, y3 = [], [], []
z1, z2, z3 = [], [], []

for i in range(len(x)):
    if(r[i] < zeros[0] + precision and r[i] > zeros[0] -precision):
        x1.append(x[i])
        y1.append(y[i])
        z1.append(z[i])
    if(r[i] < zeros[1] + precision and r[i] > zeros[1] -precision):
        x2.append(x[i])
        y2.append(y[i])
        z2.append(z[i])
    if(r[i] < zeros[2] + precision and r[i] > zeros[2] -precision):
        x3.append(x[i])
        y3.append(y[i])
        z3.append(z[i])

scatter1 = ax.scatter(x1, y1, c=z1, cmap='Reds_r')
fig.colorbar(scatter1, ax=ax, label='n')

scatter2 = ax.scatter(x2, y2, c=z2, cmap='Blues_r')
fig.colorbar(scatter2, ax=ax, label='n')

scatter3 = ax.scatter(x3, y3, c=z3, cmap='Greens_r')
fig.colorbar(scatter3, ax=ax, label='n')

plt.tight_layout()
# plt.savefig("scatter_plot.png")
plt.show()
```
